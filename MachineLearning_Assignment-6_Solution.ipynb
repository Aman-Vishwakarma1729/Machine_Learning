To calculate the Pearson correlation coefficient, we need to first calculate the covariance and standard deviation of both variables. 

Let X be the number of sales calls made per day, and Y be the number of sales made per week. We have data on both variables from a sample of 30 sales representatives. Let's denote the sample means and sample standard deviations of X and Y by:

x = sample mean of X
y = sample mean of Y
s_x = sample standard deviation of X
s_y = sample standard deviation of Y

The sample covariance between X and Y can be calculated as:

s_xy = [(X_i - x)(Y_i - y)] / (n - 1)

where  represents the sum from i=1 to n, and n is the sample size (in this case, n = 30). 

Using the sample data, we can plug in the values and calculate:

s_xy = [(X_i - x)(Y_i - y)] / (n - 1)
     = [(36-30.6)(21-19.5) + (30-30.6)(16-19.5) + ... + (43-30.6)(25-19.5)] / 29
      134.62

Next, we can calculate the sample standard deviations of X and Y using:

s_x = sqrt[(X_i - x)^2 / (n - 1)]
s_y = sqrt[(Y_i - y)^2 / (n - 1)]

Plugging in the values, we get:

s_x = sqrt[(X_i - x)^2 / (n - 1)] 
    = sqrt[((36-30.6)^2 + (30-30.6)^2 + ... + (43-30.6)^2) / 29]
     4.65

s_y = sqrt[(Y_i - y)^2 / (n - 1)] 
    = sqrt[((21-19.5)^2 + (16-19.5)^2 + ... + (25-19.5)^2) / 29]
     2.45

Now we have all the values we need to calculate the Pearson correlation coefficient:

r = s_xy / (s_x * s_y)

Plugging in the values, we get:

r = s_xy / (s_x * s_y)
   0.721

Therefore, the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week is approximately 0.721. This indicates a moderately strong positive correlation between the two variables, meaning that as the number of sales calls made per day increases, the number of sales made per week tends to increase as well.
