```python
df1['Dep_Time_hour'] = [int(i) for i in df1['Dep_Time'].str.split(':').str[0]]
df1['Dep_Time_min'] = [int(i) for i in df1['Dep_Time'].str.split(':').str[1]]
```

* We can say from here hour of jorney does not have any major affect on price

```python
Total_Num_Of_Stops = []
for i in df1['Total_Stops']:
    j = str(i).split()
    if j[0] == 'non-stop':
        Total_Num_Of_Stops.append(0)
    elif j[0] == 'nan':
        Total_Num_Of_Stops.append(1)         ## Since the mode is 1 for these column
    else:
        Total_Num_Of_Stops.append(j[0])
```

```python
df1['Total_Number_Of_Stops'] = Total_Num_Of_Stops
df1['Total_Number_Of_Stops'] = df1['Total_Number_Of_Stops'].astype(int)
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x=df1['Total_Number_Of_Stops'],y=df1['Price'])
plt.xticks(rotation = 90)
```

* We can clealry observe that number of stops doest affect the price as the number of stops increase price increases. 

```python
Duration_min = []
for i in df1['Duration']:
    try:
        i.split()[1]
    except:
        Duration_min.append('0')
    else:
        Duration_min.append(i.split()[1].removesuffix('m'))
```

```python
df1['Duration_min'] = Duration_min
df1['Duration_min'] = df1['Duration_min'].astype(int)
df1.drop(columns='Duration')
```

```python
df1['Duration_hour'] = df1['Duration'].str.split().str[0].str.removesuffix('h').str.removesuffix('m')
df1['Duration_hour'] = df1['Duration_hour'].astype(int)
```

```python
df1['Total_Duration_min'] = (df1['Duration_hour']*60)+df1['Duration_min']
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x=df1['Duration_hour'],y=df1['Price'])
plt.xticks(rotation = 90)
```

* The duration does affect the price of flight.
